# Calculus

## 1. Fungsi Linear

f(x) = ax + b


Studi Kasus: Prediksi Harga Rumah

Dalam machine learning, fungsi linear dapat digunakan untuk memprediksi harga rumah berdasarkan ukuran tanah. Misalnya, jika setiap tambahan 100 meter persegi meningkatkan harga rumah sebesar Rp50 juta, maka fungsi linear dapat direpresentasikan sebagai: f(x) = 0.5 + 200 dimana x adalah ukuran tanah dalam ratus meter persegi, dan f(x) adalah harga rumah dalam juta rupiah.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,400)
y = 2*x + 1

plt.plot(x,y, label = 'f(x) = 2x + 1')
plt.title("Linear Function")
plt.legend()
plt.show()
```

## 2. Fungsi Kuadrat

Fungsi kuadrat dinyatakan sebagai f(x) = ax**2 + bx + c (dimana tidak sama dengan 0) dan membentuk parabola.

Studi Kasus: Optimasi Biaya Produksi

Dalam industri manufaktur, biaya produksi sering kali dipengaruhi oleh jumlah produk yang diproduksi. Hubungan ini dapat dimodelkan menggunakan fungsi kuadrat: f(x) = 0.1x**2 - 5x + 100 dimana x adalah jumlah produk, dan f(x) adalah biaya produksi. Titik minimum parabola menunjukkan jumlah produk optimal untuk meminimalkan biaya.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10,10,400)
y = x**2 - 4*x + 3

plt.plot(x,y, label = 'f(x) = x**2 + 4*x + 3')
plt.title("Quadratic Function")
plt.legend()
plt.show()
```

## 3. Fungsi Polinomial

Studi Kasus: Prediksi Suhu Berdasarkan Waktu

Dalam analisis cuaca, suhu harian dapat dimodelkan menggunakan fungsi polinomial untuk menangkap tren nonlinear: f(x) = 0.1x**3 - 0.5*x**2 + 2*x -1
dimana x adalah waktu dalam jam, dan f(x) adalah suhu dalam derajat Celsius.

```python
x = np.linspace(-10, 10, 400)
y = 0.1 * x **3 - 0.5 * x**2 + 2*x - 1

plt.plot(x,y, label = 'f(x) = 0.1 * x **3 - 0.5 * x**2 + 2*x - 1')
plt.title('Polynomial Function')
plt.legend()
plt.show()
```

## 4. Fungsi Eksponensial

Fungsi eksponensial didefinisikan sebagai f(x) = a**x Sering digunakan untuk pemodelan pertumbuhan atau peluruhan.

Studi Kasus: Pertumbuhan Populasi Bakteri

Dalam biologi, pertumbuhan populasi bakteri dapat dimodelkan menggunakan fungsi eksponensial: f(x) = 2**x dimana di mana x adalah waktu dalam jam, dan f(x) adalah jumlah bakteri

```python
x = np.linspace(-10, 10, 400)  #(sumbu x awal, sumbu x akhir, jumlah titik)
y = 2**x

plt.plot(x,y, label = 'f(x) = 2**x')
plt.title('Exponential Function')
plt.legend()
plt.show()
```

## 5. Fungsi Logaritmik

Studi Kasus: Penskalaan Data Sensor

Dalam aplikasi IoT, data sensor sering kali memiliki rentang nilai yang sangat besar. Fungsi logaritmik dapat digunakan untuk menormalisasi data: f(x) = log(x) dimana x adalah nilai sensor, dan f(x) adalah nilai yang dinormalisasi.

```python
x = np.linspace(-10,10,400)
y = np.log(x)

plt.plot(x,y, label = ' f(x) = log(x)')
plt.title('Logarithmic Function')
plt.legend()
plt.show()
```

## 6. Fungsi Trigonometrik

Studi Kasus: Analisis Gelombang Suara

Dalam pemrosesan audio, gelombang suara dapat dimodelkan menggunakan fungsi sinusoidal: f(x) = sin(x) dimana x adalah waktu dalam detik, dan f(x) adalah ampitudo gelombang suara.

```python
x = np.linspace(-10,10,400)
y = np.sin(x)

plt.plot(x, y, label = 'f(x) = sin(x)')
plt.title('Sinusoidal Function')
plt.legend()
plt.show()
```

# Derivative

- Turunan pertama: Menunjukkan laju perubahan biaya terhadap jumlah produk. Misalnya, titik nol dari 
f′(x) menunjukkan titik kritis, yaitu kemungkinan minimum atau maksimum.
- Turunan kedua: Digunakan untuk memastikan apakah titik kritis tersebut adalah minimum (jika 
f′′(x)>0 atau maksimum (jika f′′(x)<0).

```python
def f(x):
    return 3 * x**3 - 6 * x**2 + x -4
def first_derivative(x):
    return 9 * x**2 - 12*x + 1
def second_derivative(x):
    return 18 * x - 12

x = np.linspace(-2,4,400)

plt.plot(x,f(x), label = 'f(x)')
plt.plot(x,first_derivative(x), label = 'first_derivative')
plt.plot(x, second_derivative(x), label = 'second_derivative')
plt.title('Derivative')
plt.legend()
plt.show()
```

# Integral

```python
def f(x):
    return x
def F(x):
    return x**2 / 2

x = np.linspace(0,2,400)
y = f(x)
F_x = F(x)

fig, ax = plt.subplots()
ax.plot(x,y, 'r', linewidth = 2, label = 'f(x) = x')
ax.plot(x, F_x, 'g', linewidth = 2, label = 'F(x) = x**2 / 2')

ax.fill_between(x,y, where = [(xi >= 0) and (xi <= 1) for xi in x], color = 'skyblue', alpha = 0.5)
ax.set_title('Integral Visualization')
ax.legend()
ax.grid(True)

plt.show()
```

# Gradient Descent

```python
def f(x):
    return x**2 - 4*x + 3

def df_dx(x):
    return 2*x - 4

def gradient_descent(init_x, learning_rate=0.1, iterations=10):
    x = init_x
    x_history = [x]
    for _ in range(iterations):
        grad = df_dx(x)
        x -= learning_rate * grad
        x_history.append(x)
    return x, x_history

x_min, x_history = gradient_descent(0.0)
x = np.linspace(-2, 4, 400)
y = f(x)

plt.plot(x, y, 'r-', label='f(x) = x^2 - 4x + 3')
plt.plot(x_history, f(np.array(x_history)), 'bo', label='Gradient Descent', markersize=3)
plt.title("Gradient Descent Visualization")
plt.legend()
plt.show()
```
