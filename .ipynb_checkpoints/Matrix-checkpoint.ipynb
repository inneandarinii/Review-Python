### Matriks


Matriks: Struktur data 2 dimensi.
Matriks diatas dibaca matriks 3 x 4 karena memiliki 3 baris dan 4 kolom.

```python
#Matriks Python (2 baris 3 kolom)
A = [[1,3,5],
     [-3,6,8]]
A
```

```python
#Matriks Bersarang
B = [[1,4,5,12], ##indeks 0
     [-5,8,9,0], ##indeks 1
     [-6,7,11,19]] ##indeks 2

print("B =", B) #Menjalankan semua baris
print("B[1] = ", B[1]) #baris kedua
print("B[2] = ", B[2]) #baris ketiga
print("B[1][2] =", B[1][2]) # B[1] = baris kedua, diambil [2] = elemen ke 3
print("B[0][-1] =", B[0][-1]) #B[0] = baris pertama, diambil [-1] elemen terakhir pertama dari belakang

column =[]
for row in B:
    column.append(row[2])

print("3rd column =",column)
```

## Membuat Numpy Array

```python
import numpy as np

#membat vektor
a = np.array([1,2,3,4,5])
a

b = np.array([1.5 , 2.5 , 5, 6, 7]) #koma dengan titik
b

#Membuat vektor dengan range
#c = np.arrange(batas bawah, batas atas, range loncat)
c = np.arange(1,10,2)
c

#Membuat linespace
d = np.linspace(1,10,4) #memnampilkan 4 angka dengan jarak yang sama dalam range 1 sampai 10.
d

#Array Multidimensi / matriks
e = np.array([(1,2,3), (4,5,6)])
e
#matriks skalar +1, maka semua komponennya ditambah 1
e + 1

#Matriks dengan nilai nol
f = np.zeros((5,5)) #matriks 5 x 5
f

#matriks dengan nilai 1
g = np.ones((5,5))
g

#Matriks identitas
h1 = np.identity(5)
h2 = np.eye(5)
print(h2)
```

## Operasi Aritmatika

```python
import numpy as np

#List python
a = [1,2,3,4,5]
b = [6,7,8,9,10]

#array np
anp = np.array ([1,2,3,4,5])
bnp = np.array([6,7,8,9,10])


#Penjumlahan
hasil = a + b
print(hasil)

#Elemen WISE operation
#Penjumlahan
hasil1 = anp + bnp
hasil1

#Pengurangan
hasil2 = anp - bnp
hasil2

#Perkalian
hasil3 = anp * bnp
hasil3

#Pembagian
hasil4 = anp / bnp
hasil4

#kuadrat
hasil5 = anp**2
hasil5
```

```python
##Multidimensi array numpy
c = np.array(([1,2,3],[4,5,6]))
d = np.array(([7,8,9],[-1,-2,-3]))

hasil = c + d
hasil

hasil1 = c *d
hasil1
```

## Indexing, Slicing, dan Iterasi

```python
import numpy as np
a = np.arange(10)**2 #arrange(10) = range 0 hingga 9 dengan dikuadratkan
print(a)

#Mengambil nilai
print("elemen ke 1 dari a adalah", a[0])
print("elemen ke 7 dari a adalah", a[6])
print("elemen keakhir dari a adalah", a[-1])

#Slicing
print("elemen dari 1 hingga 6 adalah", a[0:5]) #start_end
print("elemen dari 4 hingga akhir", a[3:])
print("elemen dari awal sampai 5", a[:5])

#Iterasi
for i in a:
    print("value=", i)
```

## Perkalian Matrix

```python
import numpy as np

#matrix multidimensi
a = np.array(([1,2],[3,4]))

#matrix satuan
b = np.ones([2,2])

print("matrix a= ")
print(a)

print("matrix b= ")
print(b)

#perkalian matrix
c1 = np.dot(a,b)
c2 = a.dot(b) # np.dot(ab) == a.dot(b)

print("matrix c1=")
print(c1)

print("matrix c2=")
print(c2)
```

## Manipulasi Matrix

```python
import numpy as np

a = np.array((
            [1,2,3],
            [4,5,6]
            ))

print("Matrix a dengan ukuran:", a.shape)

#Transpose matrix (baris dijadikan kolom)
print("Transpose matrix dari a;")
print(a.transpose())
print(np.transpose(a))
print(a.T)

#Flatten array, vektor baris
print("flatten matrix a:")
print(a.ravel())
print(np.ravel(a))

#reshape matrix
print("reshape matrix a:")
print(a.reshape(3,2))
print("reshape matrix a:")
print(a.reshape(6,1))

#resize matrix
print("resize matrix a:")
a.resize(3,2)
print(a)
```

## Stacking Matrix

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

#stacking matrix, menumpuk matrix
c = np.hstack((a,b)) #menumpuk matrix menjadi horizontal
c
d = np.vstack((a,b)) #menumpuk matrix menjadi vertikal
d
```

```python
aMat = np.zeros((2,3))
bMat = np.ones((2,3))

cMat = np.hstack((aMat, bMat))
cMat
dMat = np.vstack((aMat, bMat))
dMat
```
