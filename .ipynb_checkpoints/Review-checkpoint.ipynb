## Operators in Python

```python
#Penjumlahan
a = 7 + 10
print(f" a = {a}")
# Pengurangan
b = 10 - 9
print(f" b = {b}")
# Perkalian
c = 10 * 2
print(f" c = {c}")
#Pembagian
d = 4 / 2
print(f" d = {d}")
# Modulus : sisa pembagian
e = 10 % 3 
print(f" e = {e}")
```

### Operator Perbandingan

```python
#Ouput dalam bentuk boolean 
#Lebih dari >
x = 5 > 3
print(f" x = {x}")

#Sama dengan ==
y = 10 ==10
print(f" y = {y}")

#kurang dari <
z = 100 < 20
print(f" z = {z}")

#tidak sama dengan !=
a = 8 != 10
print(f" a = {a}")
```

### Operator Logika

```python
#Output dalam bentuk boolean
#AND
a = (5 > 3 ) and (5 > 2)
print(f" a = {a}")

#OR
b = (5 > 3) or (10 < 20)
print(f" b = {b}")

#NOT
c = not(4 > 10)
print(f" c = {c}")
```

## Control Flow in Python (if, elif, else)

```python
#beri pernyataan
age = 20

#pernyataan kondisional
if age > 20:
    print(f"Anda sudah dewasa!")
elif age == 20:
    print(f"Anda mulai memasuki usia dewasa.")
else:
    print(f"Anda belum dewasa")
```

```python
#for loop
fruits = ["Apel", "Anggur", "Nanas"]

for fruit in fruits:
    print(fruit)

for i in range (1,10):
    print(i, "=", i%2 ==0)
```

```python
#while loop

#pernyataan
count = 6
while count < 10:
    print(count)
    #kodisi
    count+=2
```

```python
#break 
for i in range (10):
    if i == 5:
        break
    print(i)
```

```python
#continue
for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
#pass
for i in range(10):
    pass
```

## Data Structures in Python

```python
#List (dapat diubah)
my_list = [1, "list", 3.17]
print(f"my list = {my_list}")

#Tuple (tidak dapat diubah)
my_tuple = (1, "list", 7.9)
print(f"my tuple = {my_tuple}")

#Set (menghapus elemen duplikat)
my_set = {7,7,8,8,"anggur", "anggur"}
print(f"my set = {my_set}")

#Dictionary
person = {"nama":"Inne", "umur":24, "kota": "Tangerang"}
print(f"person = {person}")

#List Comprehension (cara ringkas membuat list)
#Memasukan fungsi dalam [] list
list_1 = [x**2 for x in range(10)]
list_1
```

## Object Oriented Programming

```python
class Dog:
    def __init__(self,name):
        self.name = name

    def bark(self):
        print(f"{self.name} menggonggong!")
my_dog = Dog("Buddy")
my_dog.bark()
```
