Terdapat inner class dan outer class. Ketika ada orang melakukan penambahan atau perubahan pada outer class maka yang dipanggil akan yang berada pada outer class. Dan inner class tidak akan terpanggil.

Lalu bagaimana caranya supaya inner class terpanggil?
__ untuk private access modifier
_ untuk protected


```python
##Inner class
class Player:
    #attribute
    name = "inne"
    __health = 100

    #method
    def get_health(self):
        return self.__health

#Outer class
player = Player()
player.name = "shenzel"
player.health = 200

#print
print(player.get_health())
print(player.health)

print(f"{player.health} berhasil masuk")  #f itu format supaya lebih rapi outputnya
```

### 3. Polimorfisme dengan Duck Typing

```python
#Prinsip "duck typing" memungkinkan objek digunakan selama memiliki metode atau atribut yang diperlukan.

class Car:
    def drive(self):
        return "Mobil sedang dikendarai."

class Bicycle:
    def drive(self):
        return "Sepeda sedang dikayuh."

def start_vehicle(vehicle):
    print(vehicle.drive())

car = Car()
bicycle = Bicycle()

start_vehicle(car)
start_vehicle(bicycle)
```

## Enkapsulasi
Enkapsulasi adalah konsep yang membatasi akses langsung ke atribut atau metode suatu objek. Atribut atau metode yang sensitif dapat "dilindungi" dengan membuatnya privat (menggunakan _ atau __ di Python) dan hanya mengizinkan akses melalui metode publik (getter dan setter).

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # Atribut privat
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())
```

## Abstraksi
Abstraksi adalah penyederhanaan sistem kompleks dengan hanya menampilkan antarmuka yang diperlukan.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract Base Class
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        return "Mobil dinyalakan."

class Bike(Vehicle):
    def start(self):
        return "Sepeda dinyalakan."

car = Car()
print(car.start())
```
